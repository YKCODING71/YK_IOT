- 서로 관련된 함수, 클래스, 변수를 하나의 파일에 모아놓은 것
- 모듈을 사용하면 코드를 재 사용하고, 코드의 가독성을 높이며, 관리하기 쉽게 만든다.

- 모듈불러오기\
  --> 전체모듈 불러오기 : import 모듈이름\
  --> 부분모듈 불러오기 : from 모듈이름 import 함수이름
  

```python
#내장모듈 불러오기

import random

print('주사위를 굴립니다.')
player = random.randint(1,6)
computer = random.randint(1,6)
print(f'당신의 주사위 : {player}, 컴퓨터의 주사위 : {computer}')

if player > computer :
      print('당신이 이겼습니다.')
elif player < computer :
      print('컴퓨터가 이겼습니다.')
else :
      print('무승부 입니다.')
```

클래스를 사용하면 코드의 재 사용성이 높아지고, 프로그램 구조가 명확해짐

- 클래스 : 객체를 생성하기 위한 템플릿이나 기본틀

- 속성과 메서드 \
  --> 클래스의 속성 : 객체의 상태 정의(변수) \
  --> 클래스의 메서드 : 객체의 동작 정의(함수)

```python
#클래스 선언

class car :
  def __init__(self, color, model):
    self.color = color
    self.model = model

  def drive(self):
    print(f'{self.color} {self.model}이(가) 운전 중입니다.')
```

```python
#간단한 클래스 정의
#'Person'이라는 클래스를 정의하고, 'name', 'age'속성을 가지도록 하세요.
#이 클래스는 'introduc'라는 메서드를 가지며, 이 메서드는 이름과 나이를 출력합니다.

class Person :
  def __init__(self, name, age) :
    self.name = name
    self.age = age

  def introduce(self):
    return f'Hello, my name is {self.name} and I am {self.age} years old.'
```

```python
#1. 클래스 호출
person1 = Person('Alice', 30)
person2 = Person('Bob', 25)

print(person1.introduce())
print(person2.introduce())

#2. 속성 접근

person1.age = 31
print(person1.introduce())

#3. 사용자 입력받아 객체 생성
name = input('Enter your name :')
age = int(input('Enter your age :'))
person3 = Person(name,age)
print(person3.introduce())

#(패스) 4.리스트를 사용해 여러객체 생성
people = [Person('Charlie',40), Person('Diana',35)]
for person in people :
  print(person.introduce())
```

```python
#직사각형 클래스를 정의하시오.
#'Ractangle'이라는 클래스를 정의하고, 'width'와'height'속성을 가지도록 하시요.
#이클래스는 직사각형의 면적을 계산하는 'area'메서드와 둘레를 계산하는 'perimeter'메서드를 가집니다.

class Rectangle :
  def __init__(self, width, height) :
    self.width = width
    self.height = height

  def area(self) :
    return self.width * self.height

  def perimeter(self):
    return 2* (self.width + self.height)

rect = Rectangle(4, 5)
print(rect.area())
print(rect.perimeter())
```

```python
#(패스)
#상속을 통해 코드를 재 사용하시오.
#'Person'클래스를 상속받는 'Student'클래스를 정의

class Student(Person):
  def __init__(self, name, age, student_id):
    super().__init__(name, age)

    self.student_id = student_id

  def introduce(self):
    return f'{super().introduce()}My student ID is {self.student_id}'

student = Student('Bob', 20, 'S12345')
print(student.introduce())
```
