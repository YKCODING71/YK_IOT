#HSV
- Hue(색상) :  색상
- Saturation(채도) : 색의 선명도 (100 낮은 채도, 255 매우선명)
- Value(명도) : 색의 밝기 (100 적당히 밝은 정도, 255 매우 밝은 정도)

image.png

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/dog.jpg')
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) #색,채도,명도

lower_red = np.array([170, 100, 100])
upper_red = np.array([180, 255, 255])

mask = cv2.inRange(hsv, lower_red, upper_red)
result = cv2.bitwise_and(img,img, mask = mask) #(원본, 비교대상, 특정색깔로 마스킹된 이미지)

plt.imshow(img_rgb)
plt.show()

plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()

contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

for contour in contours:
    (x, y), radius = cv2.minEnclosingCircle(contour) #원의 중심과 반지름값 반환
    cv2.circle(img_rgb, (int(x), int(y)), int(radius), (0, 255, 0), 3)

plt.imshow(img_rgb)
plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/sample2.jpg')

img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) #색,채도,명도

# lower_red = np.array([0, 100, 100])
# upper_red = np.array([10, 255, 255])
# mask = cv2.inRange(hsv, lower_red, upper_red)

# lower_green = np.array([57, 100, 100])
# upper_green = np.array([70, 255, 255])
# mask = cv2.inRange(hsv, lower_green,upper_green)
lower_blue = np.array([110, 100, 100])
upper_blue = np.array([130, 255, 255])
mask = cv2.inRange(hsv, lower_blue, upper_blue)

result = cv2.bitwise_and(img,img, mask = mask) #(원본, 비교대상, 특정색깔로 마스킹된 이미지)


contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
for contour in contours:
    (x, y), radius = cv2.minEnclosingCircle(contour) #원의 중심과 반지름값 반환
    cv2.circle(img_rgb, (int(x), int(y)), int(radius), (255, 0, 0), 3)

    image_width = img.shape[1]             # 이미지 폭
    scaled_x = int(x / image_width *180)   # 전체 이미지 폭에 대한 비율 (0~180)까지 표현

    #print(image_width, x, scaled_x)
    print(f'가로좌표값 : {scaled_x}')

plt.imshow(img_rgb)
plt.show()
```
